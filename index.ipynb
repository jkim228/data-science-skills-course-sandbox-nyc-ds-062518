```python
for n in range(1,101):
    if (n>30 and n<45 and n%8==0):
        print(str(n) + " this happens twice!")
    elif (n%5 == 0 and n%2 == 1): #divisible by 5 but not even
        print(str(n) + ' high five!')
    elif (n%2 == 1): #this is odd
        print(str(n)+ " is odd")
    elif (n%7 == 0): #this is even and divisible by 7
        print(str(n)+" is divisible by 14") 
    else: 
        print(n)
```
